```python
import matplotlib.pyplot as plt
import seaborn as sns

x = [200, 400, 600, 800, 1000] # File size in Mb
y1 = [12, 20, 34, 45, 55]      # Time taken by C in seconds
y2 = [15, 25, 36, 50, 60]      # Time taken by C++ in seconds
y3 = [18, 30, 40, 55, 70]      # Time taken by Java in seconds
y4 = [20, 35, 45, 60, 80]      # Time taken by R in seconds
y5 = [25, 40, 53, 75, 100]     # Time taken by Python in seconds

Plot = sns.lineplot(x = x, y = y1, label = "C", linewidth = "3");
sns.lineplot(x = x, y = y2, label = "C++");
sns.lineplot(x = x, y = y3, label = "Java");
sns.lineplot(x = x, y = y4, label = "R");
sns.lineplot(x = x, y = y5, label = "Python", linewidth = "3");

Plot.set_xlabel("Size of file (in Mb)")
Plot.set_ylabel("Time taken (in Sec)")
Plot.set_title("Performance of languages")
```
